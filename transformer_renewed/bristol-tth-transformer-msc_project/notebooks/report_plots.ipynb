```python
# ...existing code...
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths
dileptonic_path = '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/TTTo2L2Nu.parquet'
semileptonic_path = '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/TTToSemiLeptonic.parquet'
hadronic_path = '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/TTToHadronic.parquet'
ttH_path = '/dice/projects/CMS/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/ttH_HToInvisible_M125.parquet'

zjets_individual = [
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/ZJetsToNuNu_HT-100To200.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/ZJetsToNuNu_HT-200To400.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/ZJetsToNuNu_HT-400To600.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/ZJetsToNuNu_HT-600To800.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/ZJetsToNuNu_HT-800To1200.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/ZJetsToNuNu_HT-1200To2500.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/ZJetsToNuNu_HT-2500ToInf.parquet',
]

wjets_individual = [
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/WJetsToLNu_HT-100To200.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/WJetsToLNu_HT-200To400.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/WJetsToLNu_HT-400To600.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/WJetsToLNu_HT-600To800.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/WJetsToLNu_HT-800To1200.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/WJetsToLNu_HT-1200To2500.parquet',
    '/dice/projects/CMS/Hinv/ml_datasets_ul_241111/ml_inputs_UL2018/WJetsToLNu_HT-2500ToInf.parquet',
]

# Define target luminosity (in fb^-1)
target_luminosity = 200  # Example value for Run 2

# Load DataFrames
dileptonic = pd.read_parquet(dileptonic_path)
semileptonic = pd.read_parquet(semileptonic_path)
hadronic = pd.read_parquet(hadronic_path)
ttH = pd.read_parquet(ttH_path)

# Filter DataFrames
dileptonic = dileptonic[(dileptonic['region'] == 0) | (dileptonic['region'] == 6)]
semileptonic = semileptonic[(semileptonic['region'] == 0) | (semileptonic['region'] == 6)]
hadronic = hadronic[(hadronic['region'] == 0) | (hadronic['region'] == 6)]
ttH = ttH[(ttH['region'] == 0) | (ttH['region'] == 6)]

# Load and concatenate Z+Jets DataFrames
z_jets_dfs = [pd.read_parquet(file) for file in zjets_individual]
z_jets = pd.concat(z_jets_dfs, ignore_index=True)
z_jets = z_jets[(z_jets['region'] == 0) | (z_jets['region'] == 6)]

# Load and concatenate W+Jets DataFrames
w_jets_dfs = [pd.read_parquet(file) for file in wjets_individual]
w_jets = pd.concat(w_jets_dfs, ignore_index=True)
w_jets = w_jets[(w_jets['region'] == 0) | (w_jets['region'] == 6)]

# Scale weights to target luminosity
def scale_weights(df, target_luminosity):
    # Assuming the weight is normalized to 1 fb^-1
    df['weight_scaled'] = df['weight_nominal'] * target_luminosity
    return df

dileptonic = scale_weights(dileptonic, target_luminosity)
semileptonic = scale_weights(semileptonic, target_luminosity)
hadronic = scale_weights(hadronic, target_luminosity)
ttH = scale_weights(ttH, target_luminosity)
z_jets = scale_weights(z_jets, target_luminosity)
w_jets = scale_weights(w_jets, target_luminosity)

# Calculate sum of scaled weights for each DataFrame
sum_dileptonic = dileptonic['weight_scaled'].sum()
sum_semileptonic = semileptonic['weight_scaled'].sum()
sum_hadronic = hadronic['weight_scaled'].sum()
sum_ttH = ttH['weight_scaled'].sum()
sum_z_jets = z_jets['weight_scaled'].sum()
sum_w_jets = w_jets['weight_scaled'].sum()

# Store lengths of DataFrames
data = {
    'DataFrame': ['dileptonic', 'semileptonic', 'hadronic', 'ttH', 'z_jets', 'w_jets'],
    'Sum_Scaled_Weights': [sum_dileptonic, sum_semileptonic, sum_hadronic, sum_ttH, sum_z_jets, sum_w_jets]
}

df = pd.DataFrame(data)

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['DataFrame'], df['Sum_Scaled_Weights'], color='skyblue')
plt.xlabel('DataFrame')
plt.ylabel('Sum of Scaled Weights')
plt.title('Sum of Scaled Weights for Each DataFrame')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
# ...existing code...
```
