```python
import numpy as np
import matplotlib.pyplot as plt


# Load the first ROC data saved in an npz file.\

roc_data1 = np.load('roc_data_TransformerBinary.npz')
fpr1 = roc_data1['fpr']
tpr1 = roc_data1['tpr']
auc1 = roc_data1['auc']
# Load the second ROC data saved in another npz file.
roc_data2 = np.load('roc_data_MLPbinary.npz')
fpr2 = roc_data2['fpr']
tpr2 = roc_data2['tpr']
auc2 = roc_data2['auc']


# Plot both ROC curves on one canvas
plt.figure(figsize=(8,8))
plt.plot(fpr2, tpr2, lw=2, label=f'MLP (AUC = {auc2:.3f})')
plt.plot(fpr1, tpr1, lw=2, label=f'Transformer (AUC = {auc1:.3f})')

plt.plot([0,1], [0,1], lw=2, linestyle='--', color='gray', label='Random Chance')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
#plt.title('Comparison of ROC Curves')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
